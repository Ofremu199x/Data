```python
# import the pandas module
import pandas as pd
```

```python
# load Csv file

df = pd.read_csv('/content/NationalNames.csv')
```

```python
# show data

df.head()
```

```python
# show all male gender names and get the basic statistics

male_names = df[df['Gender'] == 'M']
male_names.describe()
```

```python
# let see the median count of male gender

male_names['Count'].median()
```

```python
# let see the mean count of male gender
male_names['Count'].mean()
```

```python
# let see the median year of male gender

male_names['Year'].median()
```

```python
# let see the mean year of male gender
male_names['Year'].max()
```

```python
# now let see gender that is female and do the basic statistics

female_names = df[df['Gender'] == 'F']
female_names.describe()
```

```python
# let see how many Female are name Elizabeth

Elizabeth_names = female_names[female_names['Name'] == 'Elizabeth']
Elizabeth_names.describe()
```
